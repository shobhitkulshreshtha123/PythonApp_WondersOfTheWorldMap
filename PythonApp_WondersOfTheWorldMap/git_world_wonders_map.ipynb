```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium as Folium
import geopy as gp
```

```python
wonder_list = ["Great Wall of China", "Petra", "Christ the Redeemer", "Machu Picchu", "Chichen Itza", 
               "Colosseum", "Taj Mahal", "Great Pyramid of Giza"]
lat= []
lon= []

app= gp.Nominatim(user_agent= 'My_Practice')
for wonder in wonder_list:
    location= app.geocode(wonder).raw
    lat.append(float(location['lat']))
    lon.append(float(location['lon']))

lat_lon= tuple(zip(lat, lon))

data_dict= {}

for i in range(len(wonder_list)):
    data_dict [wonder_list[i]]= lat_lon[i]
    
wonder_map= Folium.Map(location= [20,0], zoom_start= 1.5, tiles= "Stamen Terrain")

fgw= Folium.FeatureGroup(name= "Wonders")

for i in data_dict:
    fgw.add_child(Folium.Marker(location= data_dict[i], popup=data_dict[i], tooltip= i))
wonder_map.add_child(fgw)

```
